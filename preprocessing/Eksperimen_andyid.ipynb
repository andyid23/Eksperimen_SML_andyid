# **3. Memuat Dataset**

Dataset dimuat dari folder `namadataset_raw/`. Pastikan file dataset sudah ditempatkan di folder tersebut.

```python
# Ubah 'nama_file_dataset.csv' dengan nama file dataset Anda
raw_data_dir = 'namadataset_raw'
dataset_filename = 'nama_file_dataset.csv'
input_path = os.path.join(raw_data_dir, dataset_filename)

df = pd.read_csv(input_path)
print(f"Dataset berhasil dimuat. Shape: {df.shape}")
df.head()
```

# **4. Exploratory Data Analysis (EDA)**

Pada tahap ini, Anda akan melakukan **Exploratory Data Analysis (EDA)** untuk memahami karakteristik dataset.

```python
# Informasi dataset
print("=== Informasi Dataset ===")
df.info()
print("\n=== Statistik Deskriptif ===")
df.describe(include='all')
```

```python
# Cek missing values
print("=== Missing Values ===")
print(df.isnull().sum())

# Cek duplikasi
print(f"\nJumlah data duplikat: {df.duplicated().sum()}")

# Visualisasi distribusi fitur numerik
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
if len(numeric_cols) > 0:
    df[numeric_cols].hist(figsize=(12, 8), bins=30)
    plt.tight_layout()
    plt.show()
```

# **5. Data Preprocessing**

Pada tahap ini, data preprocessing adalah langkah penting untuk memastikan kualitas data sebelum digunakan dalam model machine learning.

```python
# --- Contoh Preprocessing ---
# 1. Handling missing values
for col in df.columns:
    if df[col].dtype in ['int64', 'float64']:
        df[col].fillna(df[col].median(), inplace=True)
    else:
        df[col].fillna(df[col].mode()[0] if not df[col].mode().empty else 'unknown', inplace=True)

# 2. Hapus duplikat
df.drop_duplicates(inplace=True)

# 3. Encoding fitur kategorikal (jika ada)
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = pd.factorize(df[col])[0]

print(f"Shape setelah preprocessing: {df.shape}")
df.head()
```

```python
# Simpan hasil preprocessing ke folder namadataset_preprocessing/
output_dir = 'namadataset_preprocessing'
os.makedirs(output_dir, exist_ok=True)
output_path = os.path.join(output_dir, 'processed_data.csv')
df.to_csv(output_path, index=False)
print(f"Data preprocessing selesai. Disimpan ke: {output_path}")
```
